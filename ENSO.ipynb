```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas
sns.set_theme(style="whitegrid")
nino34 = pandas.read_csv("PMIP_nino34_stddev_pandas.csv")
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax=sns.violinplot(data=nino34, x="generation", y="nino34-stddev", hue="experiment",
               split=True, inner="quart", linewidth=1,
               palette={"piControl": "b", "midHolocene": ".85"})
ax=sns.swarmplot(x="generation", y="nino34-stddev", hue="experiment",
                   data=nino34, palette="Set2",)
```

```python
fig.savefig('ENSO.pdf')
```
